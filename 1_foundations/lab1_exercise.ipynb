```python
from dotenv import load_dotenv
load_dotenv(override=True)
import os
openai_api_key = os.getenv("OPENAI_API_KEY")
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
from openai import OpenAI
openai_client = OpenAI(api_key=openai_api_key)
messages = [
    {"role": "system", "content": "You are a helpful assistant that can answer questions and help with tasks."},
    {"role": "user", "content": "Pick a business area that might be worth exploring for an Agentic AI opportunity"}
]
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    max_tokens=100
)
business_area = response.choices[0].message.content

messages = [
    {"role": "system", "content": "You are a helpful assistant that will help with the following tasks."},
    {"role": "user", "content": f"Find pain-point in this business area: {business_area} - something challenging that might be ripe for an Agentic solution"}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    max_tokens=100
)
pain_point = response.choices[0].message.content

messages = [
    {"role": "system", "content": "You are a helpful assistant that will help with the following tasks."},
    {"role": "user", "content": f"Find a solution to this pain-point: {pain_point} - something that might be ripe for an Agentic solution"}
]

response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
solution = response.choices[0].message.content

print(solution)




```
