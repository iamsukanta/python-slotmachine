```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv("OPENAI_API_KEY")
gemini_api_key = os.getenv("GEMINI_API_KEY")

if openai_api_key:
    print(f"OpenAI API key found: {openai_api_key[:8]}")
else:
    print("OpenAI API key not found")

if gemini_api_key:
    print(f"Gemini API key found: {gemini_api_key[:8]}")
else:
    print("Gemini API key not found")
```

```python
# openai call
openai_client = OpenAI(api_key=openai_api_key)
message = [{ 'role': 'user', 'content': 'When Berlin Wall was constructed, what was the consequence for the city of Berlin?'}];
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=message
)
print(response.choices[0].message.content)



```

```python
gemini_client = OpenAI(api_key=gemini_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai")

message = [{ 'role': 'user', 'content': 'If I get into skilled employment visa not blue card how many months it will take to apply PR if I have B1 in Germay?'}]

response = gemini_client.chat.completions.create(
    model='gemini-2.5-flash',
    message= message
)

print(response.choices[0].message.content)
```
